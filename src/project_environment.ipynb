## Create Benchmark Environments

### Imports

```python
from shapely.geometry import Polygon
from shapely import plotting
import matplotlib.pyplot as plt
from IPEnvironmentShapeRobot import *
```

### Constants

```python
X_LIMITS = [0.0, 20.0]
Y_LIMITS = [0.0, 20.0]
```

### Function to visualizes Environments

```python
def create_axes(limits: list =[X_LIMITS, Y_LIMITS], figsize: tuple = (10,10)) -> plt.Axes:
    _, ax = plt.subplots(figsize=figsize)
    ax.set_xlim(limits[0][0], limits[0][1])
    ax.set_ylim(limits[1][0], limits[1][1])
    ax.set_xticks(range(0, int(limits[0][1])))
    ax.set_yticks(range(0, int(limits[1][1])))
    ax.grid()
    return ax

def draw_enviroments(env: dict, limits: list = [X_LIMITS, Y_LIMITS], ax: plt.Axes = None) -> plt.Axes:
    if not ax:
        ax = create_axes()
    for key, value in env.items():
        plotting.plot_polygon(value, add_points=False, color='red', ax=ax)

    return ax

def draw_path(path_positions: list, ax: plt.Axes = None) -> plt.Axes:
    if not ax:
        ax = create_axes()
    
    # Draw a line connecting the path nodes
    x_val, y_val = zip(*path_positions)
    ax.plot(x_val, y_val, color="k")
    ax.scatter(x_val[1:-1], y_val[1:-1], color="k", marker="o")

    # Draw Start and Goal in different colors
    # Start
    ax.scatter(x_val[0], y_val[0], marker="o", color="lightgreen", s=300)
    ax.text(x_val[0], y_val[0], "S", fontweight="heavy", size=16, ha="center", va="center")

    # Goal
    ax.scatter(x_val[-1], y_val[-1], marker="o", color="lightblue", s=300)
    ax.text(x_val[-1], y_val[-1], "G", fontweight="heavy", size=16, ha="center", va="center")

    return ax
```

### Environment 1

```python
env_1 = dict()
env_1["obs_1"] = Polygon([(5.0, 0.0), (8.0, 0.0), (8.0, 9.0), (5.0, 9.0)])
env_1["obs_2"] = Polygon([(11.0, 0.0), (16.0, 0.0), (16.0, 6.0), (11.0, 6.0)])
env_1["obs_3"] = Polygon([(10.0, 8.0), (15.0, 8.0), (15.0, 20.0), (10.0, 20.0)])

start = (2.0, 2.0)
goal = (18.0, 2.0)

solution_positions = [start, (4.0, 10.0), (8.0, 10.0), (9.0, 7.0), (16.0, 7.0), goal]

ax = create_axes(figsize=(10, 10))
draw_enviroments(env_1, ax=ax)
draw_path(solution_positions, ax=ax)
```
